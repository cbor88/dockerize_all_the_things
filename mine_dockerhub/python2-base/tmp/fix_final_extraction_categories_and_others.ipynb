```python
import csv

import numpy as np
import pandas as pd

SITE_NAME = 'megamamute'

FILE = 'files/csvs/{}.extraction.csv.SEND_TO_GDRIVE.csv'.format(SITE_NAME)
SEPARATOR = '\t'
DEPARTMENT_SEPARATOR = ','

columns_dtypes = {
 'sku': np.str,
 'ean': np.str,
 'brand': np.str,
 'name': np.str, 
 'product_type': np.str,
 'price': np.float,
 'special_price': np.float,
 'department': np.str,
 'category': np.str,
 'url': np.str,
 'available': np.bool,
 'instalments': np.int32,
 'instalment_value': np.float,
 'image_url': np.str,
 'variation_label' : np.str,
 'tyre.width': np.int32,
 'tyre.aspect_ratio': np.int32,
 'tyre.rim_diameter': np.int32,
 'tyre.load_index': np.int32,
 'tyre.speed_rating': np.str
}

df = pd.read_csv(FILE, SEPARATOR, error_bad_lines=False, dtype=columns_dtypes)

blank_fields_values = {
 'sku': 0,
 'ean': '',
 'brand': '',
 'name': '', 
 'product_type': '',
 'price': 0.0,
 'special_price': 0.0,
 'department': '',
 'category': '',
 'url': '',
 'available': True,
 'instalments': 0,
 'instalment_value': 0.0,
 'image_url': '',
 'variation_label' : '',
 'tyre.width': 0,
 'tyre.aspect_ratio': 0,
 'tyre.rim_diameter': 0,
 'tyre.load_index': 0,
 'tyre.speed_rating': ''
}
df = df.fillna(value=blank_fields_values)
```

### Below, for each row, I break the value of the column "department" into 3 different fields, in the following order: department, category, product_type:

```python
def get_department(row):
    splitted = row['department'].split(DEPARTMENT_SEPARATOR)
    return splitted[0].replace('|','').strip()

def get_category(row):
    splitted = row['department'].split(DEPARTMENT_SEPARATOR)
    if len(splitted) >= 2:
        return splitted[1].replace('|','').strip()
    else:
        return ''
    
def get_product_type(row):
    splitted = row['department'].split(DEPARTMENT_SEPARATOR)
    if len(splitted) >= 3:
        return splitted[2].replace('|','').strip()
    else:
        return ''    

df['category'] = df.apply(get_category, axis=1)  # axis = 1 applies the function to each row
df['product_type'] = df.apply(get_product_type, axis=1)
# since category and product type already are fed, now I 
# can overwrite department
df['department'] = df.apply(get_department, axis=1)

df.head(3)
```

### Format the floats with 2 decimal places

```python
df['price'] = df['price'].map(lambda price: np.around(price, decimals=2))
df['special_price'] = df['special_price'].map(lambda special_price: np.around(special_price, decimals=2))
df['instalment_value'] = df['instalment_value'].map(lambda instalment_value: np.around(instalment_value, decimals=2))
```

### Calculate the instalment_value if blank

```python
def calculate(row):
    if row['instalment_value'] > 0.0:
        return row['instalment_value']
    
    instalment_value = 0.0
    reference = row['special_price']
    instalments = row['instalments']
    if (reference > 0.0) and (instalments > 0):
        instalment_value = round(reference / instalments, 2)
    return instalment_value
 
df['instalment_value'] = df.apply(calculate, axis=1)
```

```python
df = df.rename(columns={'sku':'_id.id'})
```
