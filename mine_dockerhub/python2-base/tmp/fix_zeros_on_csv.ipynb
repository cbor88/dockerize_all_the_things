```python
import csv

import numpy as np
import pandas as pd

FILE = 'files/csvs/magazineLuiza_ean.csv'
SEPARATOR = ';'

df = pd.read_csv(FILE, SEPARATOR, error_bad_lines=False, dtype=np.str)
df = df.fillna(value='')
```

### Below, for each row, I fix the SKU to have six digits

```python
def get_sku(row):
    sku = row['SKU'].strip()
    if len(sku) < 7:
        return '0' + sku
    else:
        return sku

df['SKU'] = df.apply(get_sku, axis=1)  # axis = 1 applies the function to each row

df.tail(5)
```

```python
df.to_csv('{}.FINAL.csv'.format(FILE), sep=';', orient='records', index=False)
```
