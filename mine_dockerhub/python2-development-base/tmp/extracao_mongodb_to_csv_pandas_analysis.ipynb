## Validation of the extracted data

## "Perl" validation 

- Deve conter name e sku;
- Produto sem qualquer um dos preços é considerado indisponível;
- Produto com flag indisponível, os preços são zerados;
- (?Crawler-Artur) Codigo de barras é descartado quando tfor < 8 digitos, > 14 digitos ou = 9 digitos;
- (?Crawler-Artur) todos os campos são removidos os espaços no início e fim da string;

```python
qtde_sem_name = len(df[df['name'].isin(['N/A'])])
qtde_sem_sku = len(df[df['_id.id'].isin(['N/A'])])
qtde_sem_price_or_special_price = len(df[(df['price'] == 0) | (df['special_price'] == 0)])
qtde_available = len(df[df['available'].isin([True])])
qtde_not_available = len(df) - qtde_available
print "SEM NAME: ", str(qtde_sem_name)
print "SEM SKU: ", str(qtde_sem_sku)
print "SEM PRICE OU SPECIAL PRICE: ", str(qtde_sem_price_or_special_price)
print u"DISPONÍVEIS: ", str(qtde_available)
print u"INDISPONÍVEIS: ", str(qtde_not_available)
```

### Products without category, department, product_type and brand

```python
qtde_sem_categoria = len(df[df['category'].isin(['N/A'])])
qtde_sem_departamento = len(df[df['department'].isin(['N/A'])])
qtde_sem_tipo_produto = len(df[df['product_type'].isin(['N/A'])])
qtde_sem_brand = len(df[df['brand'].isin(['N/A'])])
print "SEM CATEGORIA: ", str(qtde_sem_categoria)
print "SEM DEPARTAMENTO: ", str(qtde_sem_departamento)
print "SEM TIPO DE PRODUTO: ", str(qtde_sem_tipo_produto)
print "SEM BRAND: ", str(qtde_sem_brand)
qtde_com_categoria = len(df) - qtde_sem_categoria
qtde_com_brand = len(df) - qtde_sem_brand
```

```python
# SETUP PYLAB
%pylab inline
from pylab import rcParams

# Setup the graph width and height (in inches):
rcParams['figure.figsize'] = 10, 10  # width, height


import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(111)

# SETUP THE DATA TO PLOT

N = 8  # the number of bar groups

# Values of the "without" bar:

semValor = [qtde_sem_name, qtde_sem_sku, qtde_sem_price_or_special_price,
            qtde_not_available, qtde_sem_categoria, qtde_sem_departamento,
            qtde_sem_tipo_produto, qtde_sem_brand]

# Values of the "total" bar:
# DONE: Fix here, put the csv len, count or size
total = []
for item in semValor:
    total.append(total_records)


# SETUP THE PLOTTING

# the x locations for the bar groups:
ind = np.arange(N)  

# the width of the bars:
width = 0.3

# Setup each individual bar
without_bar = ax.bar(ind, semValor, width,
                color='red',
                error_kw=dict(elinewidth=2, ecolor='red'))

total_bar = ax.bar(ind + width, total, width,
                    color='black',
                    error_kw=dict(elinewidth=2, ecolor='black'))

# Setup axis and labels
ax.set_xlim(-width,len(ind) + width)


# Adds 15% to the height of the graph
additional_ylim = (total_records * 15)/100
ax.set_ylim(0, total_records + additional_ylim)
ax.set_ylabel('Quantidade de itens')

ax.set_title(u'Sumário da Extração\n')

xTickMarks = ['Sem nome', 'Sem SKU', 'Sem price/special_price',
              'Indisponiveis', 'Categorias', 'Departamentos',
              'Tipo de produto', 'Marca']
ax.set_xticks(ind + width)
xtickNames = ax.set_xticklabels(xTickMarks)
plt.setp(xtickNames, rotation=45, fontsize=10)

# Setup the legend
ax.legend( (without_bar[0], total_bar[0]), ('Sem valor/Indisponiveis', 'Totals') )

# Plot the graph:
plt.show()
```

## Specific queries

```python
# Total of products on each category
category_group = df.groupby('category')
# Notice that we use size() and not count(). This is because count() applies the function to each column.
categories_counts = category_group.size()
```

```python
# Compare the total of found categories VS the total of skus on the df
sum_categories = 0
for c in categories_counts.iteritems():
    print ">> CATEGORIES: ", repr(c)
    sum_categories = sum_categories + c[1]
    
print ">>> SUM CATEGORIES........: ", str(sum_categories)
print ">>> QTDE COM CATEGORIES...:", str(qtde_com_categoria)
    
if sum_categories != qtde_com_categoria:
    print "{} skus have categories, but the sum of skus on all categories ({}) differs.".format(
        qtde_com_categoria, sum_categories)

    
```
