# Changing the input CSV to be imported by assortment (PostgreSQL)

```python
from datetime import datetime
import os
import numpy as np
import pandas as pd

# Change here for extra, walmart, etc...
SPIDER_NAME = 'walmart'

# Consolidate vars for the necessary paths
CSV_NAME = '{}.csv'.format(SPIDER_NAME)
ROOT_PATH = 'files/extracao_produtos_sites/'
FULL_INPUT_CSV_PATH = os.path.join(ROOT_PATH, CSV_NAME)

# The transaction ID (you must get its value from the "sortimento" PHP app)
TRANSACTION_ID = 1454569403

df = pd.read_csv(FULL_INPUT_CSV_PATH)

# specify which value to assume for each column if its' contents is blank or NaN
blank_fields_values = {
    'sku': '(vazio)',
    'name': '(vazio)',
    'price': 0,
    'special_price': 0,
    'category': '(vazio)',
    'department': '(vazio)' 
}

df = df.fillna(value=blank_fields_values)

print "TOTAL RECORDS:"
df.count()

# select just the records where the price is below 90000 (to manually fix a bug that ) 
df = df[(df['price'] < 90000)]
print "TOTAL RECORDS BELOW 90000:"
df.count()
```

```python
# Show all columns
df.columns
```

```python
# Get metadata about the dataframe.
df.describe()
```

```python
# Show the first 5 records
df.head(5)
```

```python
# Add a "sku" column, with data copied from the "_id.id" column
df['sku'] = df['_id.id']

# Add a "created_at" column, with the timestamp from now - 2015-03-06 16:10:08
df['created_at'] = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
```

```python
df.head(3)
```

```python
# SORTIMENTO: table "product"
# id, transaction, sku, brand, name, model, product_type, 
# price, special_price, department, category, domain, url, ean, available, created_at

'''
  "product" table:
      sku character varying(255) NOT NULL,
      name character varying(255) NOT NULL,
      price numeric(10,2) NOT NULL,
'''

# Add the "transaction" column with a fixed value, since the field can't be null:
df['transaction'] = TRANSACTION_ID

# Put the columns on the order expected by PostgreSQL
cols = ['transaction', 'sku', 'brand', 'name', 'product_type', 'price', 'special_price', 'department', 'category', 'url', 'available', 'created_at']
assortment_df_full = df[cols]
```

```python
# Full extraction (can contain NULL fields: "sku, name, price")
assortment_df_full.to_csv('/lab/files/walmart_final.csv', index=False, sep=';')
```
