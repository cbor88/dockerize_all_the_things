# Inserting CSV data into Mongo through Pandas. 
Pandas makes the process easier with the Dataframe, and allows to make adjustments before sending to Mongo.

The process consists of: using Pandas to read the CSV, convert it into a Dataframe, get_json() from it and insert into Mongo.

## Manipulating the CSV with a Pandas Dataframe:

```python
import csv

import numpy as np
import pandas as pd

SITE_NAME = 'megamamute_perl'  # kabum_perl

FILE = 'files/csvs/{}.csv'.format(SITE_NAME)
SEPARATOR = ','

columns_dtypes = {
 '_id.id': np.str,
 'ean': np.str,
 'brand': np.str,
 'name': np.str, 
 'product_type': np.str,
 'price': np.float,
 'special_price': np.float,
 'department': np.str,
 'category': np.str,
 'url': np.str,
 'available': np.bool,
 'instalments': np.int32,
 'instalment_value': np.float,
 'image_url': np.str,
 'variation_label' : np.str,
 'tyre.width': np.int32,
 'tyre.aspect_ratio': np.int32,
 'tyre.rim_diameter': np.int32,
 'tyre.load_index': np.int32,
 'tyre.speed_rating': np.str
}

df = pd.read_csv(FILE, SEPARATOR, error_bad_lines=False, dtype=columns_dtypes)

blank_fields_values = {
 '_id.id': '',
 'ean': '',
 'brand': '',
 'name': '', 
 'product_type': '',
 'price': 0.0,
 'special_price': 0.0,
 'department': '',
 'category': '',
 'url': '',
 'available': True,
 'instalments': 0,
 'instalment_value': 0.0,
 'image_url': '',
 'variation_label' : '',
 'tyre.width': 0,
 'tyre.aspect_ratio': 0,
 'tyre.rim_diameter': 0,
 'tyre.load_index': 0,
 'tyre.speed_rating': ''
}
df = df.fillna(value=blank_fields_values)
```

## Setup the mongo connection:

```python
import datetime

from pymongo import MongoClient

# CONFIGURATION
MONGO_HOST = '172.17.42.1'  # DOCKER HOST
MONGO_PORT = 27017
MONGO_DATABASE = 'insert_test'
MONGO_COLLECTION = 'items'

# CONNECT TO THE DESIRED COLLECTION
client = MongoClient(MONGO_HOST, MONGO_PORT)
db = client[MONGO_DATABASE]
collection = db[MONGO_COLLECTION]
```

```python
import json

def iter_records_dict(records):                                                                                                                                     
    for record in records:
        new_dict = {}
        # Treats just one "." in the key
        for key, value in record.iteritems():
            if '.' not in key:
                new_dict[key] = value
            else:
                split = key.split('.')
                if split[0] not in new_dict.keys():
                    new_dict[split[0]] = {}
                new_dict[split[0]][split[1]] = value
        # return new_dict
        yield new_dict

jsons = json.loads(df.to_json(orient='records'))
```

```python
collection.insert(iter_records_dict(jsons), continue_on_error=True)
```
