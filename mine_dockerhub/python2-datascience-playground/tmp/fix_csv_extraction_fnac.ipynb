```python
'''
Here I read a dataframe, apply 2 filters generating 2 sub-dataframes, 
manipulate the filtered records and create a new dataframe 
concatenating both sub-dataframes.
'''

import csv

import numpy as np
import pandas as pd

FILE = 'files/csvs/fnac_scraped.csv.CONVERTED.csv'
SEPARATOR = '\t'

df = pd.read_csv(FILE, SEPARATOR, error_bad_lines=False)

blank_fields_values = {
 '_id.id': 0,
 'ean': '',
 'brand': '',
 'name': '', 
 'product_type': '',
 'price': 0.0,
 'special_price': 0.0,
 'department': '',
 'category': '',
 'url': '',
 'available': True,
 'instalments': 0,
 'instalment_value': 0.0,
 'image_url': '',
 'variation_label' : '',
 'tyre.width': 0,
 'tyre.aspect_ratio': 0,
 'tyre.rim_diameter': 0,
 'tyre.load_index': 0,
 'tyre.speed_rating': ''
}
df = df.fillna(value=blank_fields_values)

# df = df.fillna(value='')
```

```python
zero_prices = df[(df['price']==0.0)]
non_zero_prices = df[(df['price']>0.0)]


print 'Total ORIGINAL DF: {}'.format(len(df))
print 'Total ZERO_PRICES: {}'.format(len(zero_prices))
print 'Total NON_ZERO_PRICES: {}'.format(len(non_zero_prices))

assert len(non_zero_prices) + len(zero_prices) == len(df)
zero_prices.head(3)
```

```python
zero_prices['price'] = zero_prices['special_price']
zero_prices.head(3)
```

```python
non_zero_prices.head(3)
```

```python
# TODO: Merge both zero and non-zero dataframes
concatenated = pd.concat([zero_prices, non_zero_prices])

assert len(concatenated) == len(df)

print 'Total on the concatenated dataframe: {}'.format(len(concatenated))

concatenated.head(3)
```

```python
concatenated.to_csv('files/csvs/fnac_scraped.CONVERTED.FIXED.csv', sep='\t', orient='records', index=False)
```
