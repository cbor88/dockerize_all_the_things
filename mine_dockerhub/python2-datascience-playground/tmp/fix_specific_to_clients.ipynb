```python
import csv

import numpy as np
import pandas as pd

SITE_NAME = 'tireshop-STILL_PANDAS'

FILE = 'files/csvs/EXTRACOES_2015_AGOSTO/{}/extraction.csv.SEND_TO_GDRIVE.csv.FINAL.csv'.format(SITE_NAME)
SEPARATOR = '\t'
DEPARTMENT_SEPARATOR = ','

columns_dtypes = {
 'sku': np.str,
 'ean': np.str,
 'brand': np.str,
 'name': np.str, 
 'product_type': np.str,
 'price': np.float,
 'special_price': np.float,
 'department': np.str,
 'category': np.str,
 'url': np.str,
 'available': np.bool,
 'instalments': np.int32,
 'instalment_value': np.float,
 'image_url': np.str,
 'variation_label' : np.str,
 'tyre.width': np.int32,
 'tyre.aspect_ratio': np.int32,
 'tyre.rim_diameter': np.int32,
 'tyre.load_index': np.int32,
 'tyre.speed_rating': np.str
}

df = pd.read_csv(FILE, SEPARATOR, error_bad_lines=False, dtype=columns_dtypes)

blank_fields_values = {
 'sku': 0,
 'ean': '',
 'brand': '',
 'name': '', 
 'product_type': '',
 'price': 0.0,
 'special_price': 0.0,
 'department': '',
 'category': '',
 'url': '',
 'available': True,
 'instalments': 0,
 'instalment_value': 0.0,
 'image_url': '',
 'variation_label' : '',
 'tyre.width': 0,
 'tyre.aspect_ratio': 0,
 'tyre.rim_diameter': 0,
 'tyre.load_index': 0,
 'tyre.speed_rating': ''
}
df = df.fillna(value=blank_fields_values)
```

### Calculate the instalment_value, once I have the instalments

```python
def calculate(row):
    instalment_value = 0.0
    reference = row['special_price']
    instalments = row['instalments']
    if (reference > 0.0) and (instalments > 0) and (row['instalment_value'] <= 0.0):
        instalment_value = round(reference / instalments, 2)
    return instalment_value
 
df['instalment_value'] = df.apply(calculate, axis=1)

df.head(3)
```
