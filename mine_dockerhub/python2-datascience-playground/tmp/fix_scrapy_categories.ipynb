```python
import csv

import numpy as np
import pandas as pd

SITE_NAME = 'comercial_roma'

FILE = 'files/csvs/{}_scraped.csv'.format(SITE_NAME)
SEPARATOR = '\t'
DEPARTMENT_SEPARATOR = ','

columns_dtypes = {
 'url': np.str,
 'sku': np.str,
 'name': np.str, 
 'variation_label': np.str,
 'available': np.bool,
 'price': np.float,
 'special_price': np.float,
 'image_url': np.str,
 'brand': np.str,
 'ean': np.str, 
 'categories': np.str,
 'instalments': np.str,
 'instalment_value': np.float}

df = pd.read_csv(FILE, SEPARATOR, error_bad_lines=False, dtype=columns_dtypes)

blank_fields_values = {
 'url': '',
 'sku': '',
 'name': '', 
 'variation_label': '',
 'available': True,
 'price': 0.0,
 'special_price': 0.0,
 'image_url': '',
 'brand': '',
 'ean': '', 
 'categories': '',
 'instalments': '',
 'instalment_value': 0.0}

df = df.fillna(value=blank_fields_values)
```

### Below, for each row, I break the value of the column "categories" into 3 different fields, in the following order: department, category, product_type:

```python
def get_department(row):
    splitted = row['categories'].split(DEPARTMENT_SEPARATOR)
    return splitted[0].replace('|','').strip()

def get_category(row):
    splitted = row['categories'].split(DEPARTMENT_SEPARATOR)
    if len(splitted) >= 2:
        return splitted[1].replace('|','').strip()
    else:
        return ''
    
def get_product_type(row):
    splitted = row['categories'].split(DEPARTMENT_SEPARATOR)
    if len(splitted) >= 3:
        return splitted[2].replace('|','').strip()
    else:
        return ''    

df['category'] = df.apply(get_category, axis=1)  # axis = 1 applies the function to each row
df['product_type'] = df.apply(get_product_type, axis=1)
# since category and product type already are fed, now I 
# can overwrite department
df['department'] = df.apply(get_department, axis=1)

df.head(3)
```

```python
EXPORT_COLUMNS = ['url', 'sku', 'name', 'variation_label', 'available', 
                  'price', 'special_price', 'image_url', 'brand', 'ean',
                  'department', 'category', 'product_type', 
                  'instalments', 'instalment_value']
df.to_csv('{}.FINAL.csv'.format(FILE), sep='\t', orient='records', columns=EXPORT_COLUMNS, index=False)
```
