```python
import csv

import numpy as np
import pandas as pd

FILE = 'files/csvs/americanas.csv'
SEPARATOR = ','

df = pd.read_csv(FILE, SEPARATOR, error_bad_lines=False)
# df = pd.read_csv(FILE, SEPARATOR, error_bad_lines=False, quoting=csv.QUOTE_NONE)


blank_fields_values = {
 '_id.id': 0,
 'ean': '',
 'brand': '',
 'name': '', 
 'product_type': '',
 'price': 0,
 'special_price': 0,
 'department': '',
 'category': '',
 'url': '',
 'available': True,
 'instalments': 0,
 'instalment_value': 0,
 'image_url': '',
 'variation_label' : '',
 'tyre.width': 0,
 'tyre.aspect_ratio': 0,
 'tyre.rim_diameter': 0,
 'tyre.load_index': 0,
 'tyre.speed_rating': '',
 'possible_eans': ''
}
df = df.fillna(value=blank_fields_values)

# df = df.fillna(value='')
```
